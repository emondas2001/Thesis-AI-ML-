```python
import pandas as pd
from textblob import TextBlob
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
df = pd.read_excel("ALL_FULL_DATA MODIFIED.xlsx")
```

```python
df_300 = df.iloc[300:600].copy()
```

```python
def label_sentiment(text):
    polarity = TextBlob(str(text)).sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
df_300["Sentiment"] = df_300[""].apply(label_sentiment)
```

```python
columns_to_drop = [col for col in df_300.columns if 'Unnamed:' in col]
df_300 = df_300.drop(columns=columns_to_drop)
print(df_300.columns.tolist())
```
