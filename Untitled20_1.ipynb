```python
import os, sys, subprocess

REPO_URL = "https://github.com/EhsanulHaqueSiam/BDNewsPaperScraper.git"
PROJECT_DIR = os.path.abspath("BDNewsPaperScraper")

if not os.path.exists(PROJECT_DIR):
    subprocess.run(["git", "clone", REPO_URL, PROJECT_DIR], check=True)
else:
    subprocess.run(["git", "-C", PROJECT_DIR, "pull"], check=True)

pkgs = [
    "scrapy>=2.12.0",
    "scrapy-playwright>=0.0.42",
    "playwright>=1.49.0",
    "pandas>=2.2.0",
    "openpyxl>=3.1.0",
    "pytz",
    "python-dateutil",
]
subprocess.run([sys.executable, "-m", "pip", "install", "-q"] + pkgs, check=True)
subprocess.run([sys.executable, "-m", "playwright", "install", "--with-deps", "chromium"], check=True)

print("✅ Setup done:", PROJECT_DIR)
```

```python
import time, subprocess, shlex

spiders = [
    "prothomalo",   # NO date args
    "dailysun",     # NO date args (hardened)
    "thedailystar"  # WITH date args
]

for sp in spiders:
    print("\n" + "="*80)
    print(f"🚀 Running spider: {sp}")

    try:
        # 🔥 Daily Sun hardened (NO date args)
        if sp == "dailysun":
            cmd = f"""
            cd {shlex.quote(PROJECT_DIR)} && \
            python -m scrapy crawl dailysun \
            -L INFO \
            -s DOWNLOAD_DELAY=2 \
            -s RANDOMIZE_DOWNLOAD_DELAY=True \
            -s CONCURRENT_REQUESTS=2 \
            -s DEPTH_LIMIT=5 \
            -s AUTOTHROTTLE_ENABLED=True \
            -s AUTOTHROTTLE_START_DELAY=2 \
            -s AUTOTHROTTLE_MAX_DELAY=10 \
            -s PLAYWRIGHT_DEFAULT_NAVIGATION_TIMEOUT=60000 \
            -s USER_AGENT="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120.0.0.0 Safari/537.36"
            """

        # ✅ Daily Star supports date args
        elif sp == "thedailystar":
            cmd = (
                f"cd {shlex.quote(PROJECT_DIR)} && "
                f"python -m scrapy crawl thedailystar -L INFO "
                f"-a start_date={start_date} -a end_date={end_date}"
            )

        # 🔹 Prothom Alo (NO date args)
        else:
            cmd = (
                f"cd {shlex.quote(PROJECT_DIR)} && "
                f"python -m scrapy crawl prothomalo -L INFO"
            )

        code = subprocess.call(["bash", "-lc", cmd])

        if code != 0:
            print(f"❌ {sp} FAILED (exit code {code})")
        else:
            print(f"✅ {sp} DONE")

    except Exception as e:
        print(f"❌ {sp} CRASHED:", e)

    time.sleep(3)

print("\n🎉 DONE — ALL 3 spiders scraped data.")
```

```python
import sqlite3, pandas as pd, os

DB_PATH = os.path.join(PROJECT_DIR, "news_articles.db")

with sqlite3.connect(DB_PATH) as conn:
    df = pd.read_sql_query("SELECT * FROM articles", conn)

# 🔎 URL column auto-detect
url_col = next(c for c in df.columns if "url" in c.lower() or "link" in c.lower())
print("Using URL column:", url_col)

# 🔥 ROBUST business keywords (contains-based)
BUSINESS_KEYWORDS = [
    "prothomalo.com/business",
    "en.prothomalo.com/business",
    "daily-sun.com/business",
    "thedailystar.net/business",
]

# Normalize URL
urls = df[url_col].astype(str).str.lower()

mask = False
for kw in BUSINESS_KEYWORDS:
    mask |= urls.str.contains(kw, na=False)

finance_df = df[mask].copy()

print("✅ Total Finance / Business articles:", len(finance_df))

# ✅ Paper-wise check (IMPORTANT)
paper_counts = finance_df.groupby("paper_name").size().sort_values(ascending=False)
paper_counts
```
